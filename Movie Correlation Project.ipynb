```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'C:\Users\Justin\MySQL_Project_3\movies.csv')
```

```python
# Now lets show the read in data
df.head()
```

```python
# Let's check for missing data
    
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# find each data type

df.dtypes
```

```python
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
# Create correctyear column
df['yearcorrect'] = df['released'].str.extract('(\d{4})')
```

```python
df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# let's drop duplicates
df.drop_duplicates().head()
```

```python
# Scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])

# Plot Title
plt.title('Budget vs Gross Earnings')

# Label X
plt.xlabel('Gross Earnings')


# Label Y
plt.ylabel('Budget for Film')




plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget',y='gross',data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Let's start looking at correlation

df.corr(method = 'pearson',numeric_only=True).head() #pearson, kendall, spearman
```

```python
# High correlation between budget and gross

correlation_matrix = df.corr(method = 'pearson', numeric_only = True)
sns.heatmap(correlation_matrix, annot = True)

# Plot Title
plt.title('Correlation Matric for Numeric Features')

# Label X
plt.xlabel('Movie Features')


# Label Y
plt.ylabel('Movie Features')

plt.show()
```

```python
# Looks at company

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
# Numerized Matrix

correlation_matrix = df_numerized.corr(method = 'pearson', numeric_only = True)
sns.heatmap(correlation_matrix, annot = True)

# Plot Title
plt.title('Correlation Matric for Numeric Features')

# Label X
plt.xlabel('Movie Features')


# Label Y
plt.ylabel('Movie Features')

plt.show()
```
